```python
from transformers import AutoTokenizer, DataCollatorWithPadding
import numpy as np
import datasets
from datasets import Dataset, DatasetDict
```

```python
X_test['text'] = X_test
```

```python
X_test['labels'] = y_test
```

```python
X_test = X_test.drop(['def'], axis=1)
```

```python
dataset_test = Dataset.from_pandas(X_test)
```

```python
checkpoint = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
import transformers
from transformers import AutoTokenizer
```

```python

def tokenize_function(example):
    return tokenizer(example["text"] ,truncation=True, padding = True, max_length = 256)


tokenized_test = dataset_test.map(tokenize_function, batched=True)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer, return_tensors="tf")

tf_test_dataset = tokenized_test.to_tf_dataset(
    columns=["attention_mask", "input_ids", "token_type_ids"],
    label_cols=["labels"],
    shuffle=True,
    collate_fn=data_collator,
    batch_size=8,
)

```

```python
precision = tf.keras.metrics.Precision()
recall = tf.keras.metrics.Recall()

@tf.function
def f1_macro(y_true, y_pred):
    precision_value = precision(y_true, y_pred)
    recall_value = recall(y_true, y_pred)
    return 2 * ((precision_value * recall_value) / (precision_value + recall_value + tf.keras.backend.epsilon()))

```

```python
from transformers import TFBertForSequenceClassification
```

```python
path = '/content/drive/MyDrive/models/model.h5'
```

```python
loaded_model = tf.keras.models.load_model(path, custom_objects={'f1_macro': f1_macro, 'TFBertForSequenceClassification': TFBertForSequenceClassification})
```

```python
import tensorflow as tf
import transformers

print("TensorFlow version:", tf.__version__)
print("Transformers version:", transformers.__version__)
```

```python
loaded_model.evaluate(tf_test_dataset)
```

```python
model = loaded_model
```

```python
y_pred = model.predict(tf_test_dataset)
y_pred_binary = [int(round(p[0])) for p in y_pred]
```

```python
from sklearn.metrics import f1_score,recall_score


# Calculate recall for each class


recall_positive = recall_score(y_test, y_pred_binary, pos_label=1)
recall_negative = recall_score(y_test, y_pred_binary, pos_label=0)

print(f"Recall for positive class: {recall_positive}")
print(f"Recall for negative class: {recall_negative}")

# Display f1 score for negative and positive class and macro



# Calculate F1 score for each class
f1_negative = f1_score(y_test, y_pred_binary, average='binary', pos_label=0)
f1_positive = f1_score(y_test, y_pred_binary, average='binary', pos_label=1)

# Calculate macro F1 score
f1_macro = f1_score(y_test, y_pred_binary, average='macro')

# Print results
print("F1 score for negative class:", f1_negative)
print("F1 score for positive class:", f1_positive)
print("Macro F1 score:", f1_macro)
```
